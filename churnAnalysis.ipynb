# Hypothesis testing

To check if there's a significant difference in the number of support tickets per month between churned and non-churned customers


H0: There is no siggnicficant difference in the number of support tickets per month between churned and non-churned users.
H1: There is significant difference between the number of support tickets per month between churned and non-churned users.

```python
from scipy.stats import mannwhitneyu

# Split the data
churn_yes = data[data['Churn'] == 1]['SupportTicketsPerMonth']
churn_no = data[data['Churn'] == 0]['SupportTicketsPerMonth']

# Perform Mann-Whitney U Test
stat, p = mannwhitneyu(churn_yes, churn_no)

print('Mann-Whitney U test statistic:', stat)
print('p-value:', p)

# Interpret the result
if p < 0.05:
    print("There is a significant difference in support tickets between churned and non-churned customers.")
else:
    print(" No significant difference in support tickets between the groups.")
```

Association between Churn and SubscriptionType

H₀: Churn is independent of SubscriptionType

H₁: Churn is associated with SubscriptionType

```python
from scipy.stats import chi2_contingency
pd.crosstab(data['Churn'], data['SubscriptionType'])
chi2, p, dof, expected = chi2_contingency(pd.crosstab(data['Churn'], data['SubscriptionType']))
chi2, p, dof
```

Since the p-value (0.7497) is greater than 0.05, we fail to reject the null hypothesis.
This means there is no statistically significant association between Churn and Subscription Type.

```python
# Monthly charges vs churn
from scipy.stats import mannwhitneyu

# Split MonthlyCharges by churn
churn_0 = data[data['Churn'] == 0]['MonthlyCharges']
churn_1 = data[data['Churn'] == 1]['MonthlyCharges']

# Mann-Whitney U Test
stat, p = mannwhitneyu(churn_0, churn_1, alternative='two-sided')
print("Mann-Whitney U test for MonthlyCharges vs Churn")
print(f"Statistic: {stat}, p-value: {p}")

if p < 0.05:
    print("There is a significant difference in Monthly Charges between churned and non-churned customers.")
else:
    print("There is no significant difference in Monthly Charges between churned and non-churned customers.")
```

# Churn Prediction

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python


# Encode categorical variables (you can use get_dummies or LabelEncoder)
data_encoded = pd.get_dummies(data, drop_first=True)

# Split
X = data.drop('Churn', axis=1)
y = data['Churn']

# Combine X and y if needed to avoid index mismatch
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
model.fit(X_train_scaled, y_train)
```

```python
y_proba = model.predict_proba(X_test_scaled)[:, 1]
y_pred = (y_proba >= 0.35).astype(int)  # You can adjust threshold

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
```

The confusion matrix and classification report reveal a significant shift in model behavior due to threshold tuning and class balancing:

Confusion Matrix:

True Negatives (TN): 80

False Positives (FP): 79

False Negatives (FN): 5

True Positives (TP): 29

 Class 1 (Churners)
Recall (0.85): The model correctly identified 85% of actual churners, which is excellent and critical for churn prevention.

Precision (0.27): However, only 27% of those predicted as churners were correct. This means the model tends to flag many false positives.

F1-Score (0.41): Balances precision and recall — indicating a reasonable but not perfect model for churn detection.

 Class 0 (Non-Churners)
Recall dropped to 0.50: The model correctly identified only 50% of non-churners, sacrificing some accuracy to better capture churners.

Precision (0.94): When the model predicts non-churn, it's mostly right.

 Overall Performance
Accuracy: 56%: Lower overall accuracy due to the high number of false positives — but this is acceptable in imbalanced classification problems, where catching the minority class (churners) is more important.

#Calibrated Balanced Bagging with Logistic Regression"

```python
from imblearn.ensemble import BalancedBaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, f1_score

# STEP 2: BalancedBaggingClassifier + Calibration
bbc = BalancedBaggingClassifier(
    estimator=LogisticRegression(solver='liblinear'),
    sampling_strategy='auto',
    replacement=False,
    random_state=42
)

calibrated = CalibratedClassifierCV(bbc, method='sigmoid', cv=5)
calibrated.fit(X_train_reduced, y_train)

# STEP 3: Predict Probabilities
y_probs = calibrated.predict_proba(X_test_reduced)[:, 1]

# STEP 4: Threshold Tuning
import numpy as np

best_thresh = 0.5
best_f1 = 0

for thresh in np.arange(0.1, 0.9, 0.05):
    y_pred_thresh = (y_probs >= thresh).astype(int)
    f1 = f1_score(y_test, y_pred_thresh)
    print(f"Threshold: {thresh:.2f} → F1 Score: {f1:.4f}")
    if f1 > best_f1:
        best_f1 = f1
        best_thresh = thresh

print(f"\n Best Threshold: {best_thresh} with F1 Score: {best_f1:.4f}")


# STEP 5: Final Evaluation at Best Threshold

y_final = (y_probs >= best_thresh).astype(int)

print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_final))
print("\nClassification Report:\n", classification_report(y_test, y_final))
print("ROC AUC Score:", roc_auc_score(y_test, y_probs))
```

To improve the detection of churners (minority class), I used a Logistic Regression model with class_weight='balanced', which compensates for class imbalance by penalizing misclassification of the minority class more heavily.

After training, I applied threshold tuning by evaluating multiple probability thresholds (0.10 to 0.85) and selecting the one that gave the highest F1 Score for the positive class (churners). F1 score balances both precision and recall, making it an ideal metric when both false positives and false negatives matter.

The best threshold was 0.25, giving an F1 score of 0.46, which significantly improved recall (0.50) while maintaining good precision (0.42).

The ROC AUC score of 0.73 indicates that the model can distinguish between churners and non-churners with reasonable confidence.

Accuracy remained at 79%, showing that the improvement in churner detection didn’t come at the cost of overall performance.

This approach helps create a better business intervention strategy, allowing us to catch more potential churners without overwhelming the system with too many false positives.



```python
# STEP 1: Predict churn probabilities on test data
y_test_probs = calibrated.predict_proba(X_test_reduced)[:, 1]

# STEP 2: Apply the best threshold to get final predictions
y_test_pred = (y_test_probs >= best_thresh).astype(int)

# STEP 3: Evaluate the performance
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score

print("\n Final Evaluation on Test Data")
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_test_pred))
print("\nClassification Report:\n", classification_report(y_test, y_test_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_test_probs))
```

The model performs very well on the majority class (non-churners).

It has moderate performance on the minority class (churners) but better than a baseline or simple model.

Threshold tuning helped to improve the F1-score for churners by balancing sensitivity and specificity.
